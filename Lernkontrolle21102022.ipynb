# Step 3
### Lösche die Spalten wo die Null Werte 50%>

```python
data1=data.drop(data.columns[data.isnull().mean() >0.5], axis=1)
data1







```

# Step 4
### Prozentual gesehen, wie viel NaN Werte gibt es im ganzen DataFrame?

```python
round(data1.isnull().sum() * 100 / len(data1))




```

# Step6 (recherche Aufgabe)
## Fülle alle NaN Werte in der Spalte "Existing Use" mit dem nächsten darüber liegenden NICHT NaN Wert!
## Im zweiten Schritt, füllt die übrigen NaN Werte mit 0.
### Die beiden folgenden Tabellen sind Beispiele zur Veranschaulichung. Die erste zeigt die Ausgangstabelle und die zweite das Ergebnis nach Anwendung des zweiten Schrittes.

Spalte 1 | Spalte 2 
-------- | -------- 
NaN   | 1  
 3  | NaN 
NaN | 7  
NaN   | NaN 

##### Endergebnis
Spalte 1 | Spalte 2 
-------- | -------- 
0   | 1  
 3  | 1 
3 | 7  
3   | 7 

```python



data1['Existing Use'] = data1['Existing Use'].fillna(method='ffill')
data1

```

# Step 7
## Ersetze die NaN Werte in der Spalte "Number of Existing Stories"  mit einer 1

```python
data1[ "Number of Existing Stories" ]=data1[ "Number of Existing Stories" ].fillna(1)
data1.head(5)


```

# Step 8
## Ersetze die NaN Werte in der Spalte "Existing Units"  mit der Maximale Wert Diese Spalte

```python
data1["Existing Units"]=data1["Existing Units"].fillna(np.amax(data1["Existing Units"]))
data1.head(5)



```

# Step 9
## Ersetze die NaN Werte in der Spalte "Plansets"  mit eine Polynomial Interpolation  zweite Ordnung

```python
data1["Plansets"]=data1["Plansets"].fillna(data1["Plansets"].interpolate(method ='polynomial',order=2))
data1.head(5)




```
